```python
gt_output_dir=os.path.join(save_folder_res, "output_gt")
baseline_output_dir=os.path.join(save_folder_res, "output_baseline_high_res")
new_output_dir=os.path.join(save_folder_res, "output_new_model_high_res")

gt_paths = sorted([i for i in os.listdir(gt_output_dir) if i.startswith('frame')])
baseline_paths = sorted([i for i in os.listdir(baseline_output_dir) if i.startswith('frame')])
newmodel_paths = sorted([i for i in os.listdir(new_output_dir) if i.startswith('frame')])
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Assume these lists and directories are defined:
# baseline_paths, gt_paths, newmodel_paths, gt_output_dir, baseline_output_dir, new_output_dir, save_folder_res

nb_frames = len(baseline_paths)
i0 = 0
i1 = int(nb_frames / 4)
i2 = int(2 * nb_frames / 3)
i_last = nb_frames - 1  # Additional frame using the last frame

def cycle_title(frame_idx):
    return f"t={frame_idx/nb_frames*100:.0f}%"

# Load images for each method
# Row 1: CNN reference (formerly GT)
img_cnn0    = Image.open(os.path.join(gt_output_dir, gt_paths[i0]))
img_cnn1    = Image.open(os.path.join(gt_output_dir, gt_paths[i1]))
img_cnn2    = Image.open(os.path.join(gt_output_dir, gt_paths[i2]))
img_cnn_last = Image.open(os.path.join(gt_output_dir, gt_paths[i_last]))

# Row 2: Baseline
img_baseline0    = Image.open(os.path.join(baseline_output_dir, baseline_paths[i0]))
img_baseline1    = Image.open(os.path.join(baseline_output_dir, baseline_paths[i1]))
img_baseline2    = Image.open(os.path.join(baseline_output_dir, baseline_paths[i2]))
img_baseline_last = Image.open(os.path.join(baseline_output_dir, baseline_paths[i_last]))

# Row 3: Ours (formerly New model)
img_ours0    = Image.open(os.path.join(new_output_dir, newmodel_paths[i0]))
img_ours1    = Image.open(os.path.join(new_output_dir, newmodel_paths[i1]))
img_ours2    = Image.open(os.path.join(new_output_dir, newmodel_paths[i2]))
img_ours_last = Image.open(os.path.join(new_output_dir, newmodel_paths[i_last]))

# Create a 3x4 grid of subplots
fig, ax = plt.subplots(3, 4, figsize=(12, 8))

# First row: CNN reference
ax[0, 0].imshow(img_cnn0, cmap='gray', aspect='equal')
ax[0, 0].set_title(cycle_title(i0))
ax[0, 0].set_ylabel('CNN reference')
ax[0, 1].imshow(img_cnn1, cmap='gray', aspect='equal')
ax[0, 1].set_title(cycle_title(i1))
ax[0, 2].imshow(img_cnn2, cmap='gray', aspect='equal')
ax[0, 2].set_title(cycle_title(i2))
ax[0, 3].imshow(img_cnn_last, cmap='gray', aspect='equal')
ax[0, 3].set_title(cycle_title(i_last))

# Second row: Baseline
ax[1, 0].imshow(img_baseline0, cmap='gray', aspect='equal')
ax[1, 0].set_title(cycle_title(i0))
ax[1, 0].set_ylabel('Baseline')
ax[1, 1].imshow(img_baseline1, cmap='gray', aspect='equal')
ax[1, 1].set_title(cycle_title(i1))
ax[1, 2].imshow(img_baseline2, cmap='gray', aspect='equal')
ax[1, 2].set_title(cycle_title(i2))
ax[1, 3].imshow(img_baseline_last, cmap='gray', aspect='equal')
ax[1, 3].set_title(cycle_title(i_last))

# Third row: Ours
ax[2, 0].imshow(img_ours0, cmap='gray', aspect='equal')
ax[2, 0].set_title(cycle_title(i0))
ax[2, 0].set_ylabel('Ours')
ax[2, 1].imshow(img_ours1, cmap='gray', aspect='equal')
ax[2, 1].set_title(cycle_title(i1))
ax[2, 2].imshow(img_ours2, cmap='gray', aspect='equal')
ax[2, 2].set_title(cycle_title(i2))
ax[2, 3].imshow(img_ours_last, cmap='gray', aspect='equal')
ax[2, 3].set_title(cycle_title(i_last))

# Remove x and y ticks for a cleaner look
for a in ax.flatten():
    a.set_xticks([])
    a.set_yticks([])

# Adjust subplot spacing
plt.subplots_adjust(wspace=0.1, hspace=0.3)

# Save and display the figure
plt.savefig(os.path.join(save_folder_res, "3x4_plot.png"), bbox_inches='tight')
plt.show()
```
